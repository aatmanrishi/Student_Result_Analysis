Dropping Unnamed Columns

```python
df = df.drop('Unnamed: 0',axis=1)
print(df.head());
```

Changining  Weekly Study Column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","5-10");
print(df.head());
```

Gender Distribution 

```python
ax = sns.countplot(data=df,x="Gender");
ax.bar_label(ax.containers[0])
plt.show();
```

From the above data we have realised that the number of females is more than males 

Based on the Education of Parents we are gonna identify the students math score and writing score .

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"});
print(gb);
```

Getting Heatmap of above data

```python
sns.heatmap(gb,annot=True); #Passing ANNOT = True will add values to each color shade making chart more descriptive.
plt.title("Relationship between parent education and students score")
plt.show();
```

From the above chart we have concluded that the education of students have a critical effect on there on kids.

Let's see is there any effect of parent marital status on there kids

```python
gbMaritalStatus = gb = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"});
print(gb);
```

Let's create a heat map of above data

```python
sns.heatmap(gb,annot=True); #Passing ANNOT = True will add values to each color shade making chart more descriptive.
plt.title("Relationship between Parents Marital Status and Students Score")
plt.show();
```

From the above map we can Identify that marital status have [not that much]/Negligible impact on student education stats when compared to Parent Education .

Distribution of Ethinic Group

```python
groupA = df.loc[df['EthnicGroup']=='group A'].count();
groupB = df.loc[df['EthnicGroup']=='group B'].count();
groupC = df.loc[df['EthnicGroup']=='group C'].count();
groupD = df.loc[df['EthnicGroup']=='group D'].count();
groupE = df.loc[df['EthnicGroup']=='group E'].count();
```

```python
mlist=[groupA["EthnicGroup"],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']];

l = ['group A', 'group B' ,'group C' ,'group D' ,'group E']
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
```
